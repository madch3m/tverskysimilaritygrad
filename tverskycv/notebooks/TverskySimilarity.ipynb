Stub class for initating the tversky model for feature bank

```python
def feature_membership(self, x):
    # Compute similarity between input and feature bank
    membership_strength = torch.matmul(x, self.feature_bank.T)
    # Soft membership (smooth threshold)
    return torch.sigmoid(membership_strength)
```

Method for setting the scalars alpha and beta to ensure that they are differentiable

```python
import torch.nn.functional as F
def init_alpha(self, a):
    return F.softplus(a)
def init_beta(self, b):
    return F.softplus(b)
```
